```python
import pandas as pd

# Load the 200-row dataset
df = pd.read_csv('/content/twitter_sentiment_200.csv')

# Basic before-cleaning info
print("Rows, Columns:", df.shape)
print("\nColumns & types:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nFirst 10 rows:")
display(df.head(10))

```

```python
# Count duplicates first
dup_count = df.duplicated().sum()
print("Duplicate rows:", dup_count)

# Remove duplicates (keeps first occurrence)
df = df.drop_duplicates().reset_index(drop=True)
print("After drop_duplicates -> Rows:", df.shape[0])
```

```python
# replace 'tweet' with actual column name if needed
text_col = 'tweet' if 'tweet' in df.columns else ('text' if 'text' in df.columns else None)
print("Detected text column:", text_col)

# If no text column found, list columns and stop
if text_col is None:
    print("No 'tweet' or 'text' column found. Columns:", df.columns.tolist())
else:
    # Check missing
    print("Missing in text column:", df[text_col].isnull().sum())

    # Option A: drop rows with missing text (recommended)
    df = df.dropna(subset=[text_col]).reset_index(drop=True)

    # Option B: OR fill missing with placeholder (if you prefer)
    # df[text_col] = df[text_col].fillna("No text")
```

```python
import re

def simple_clean(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = re.sub(r'http\S+','', text)         # remove URLs
    text = re.sub(r'@\w+','', text)            # remove mentions
    text = re.sub(r'#','', text)               # remove hash sign (keep word)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text) # remove punctuation
    text = text.lower().strip()
    text = re.sub(r'\s+', ' ', text)           # collapse multiple spaces
    return text

if text_col:
    df['clean_text'] = df[text_col].apply(simple_clean)
    display(df[[text_col, 'clean_text']].head(8))
```

```python
# Example for retweet_count
if 'retweet_count' in df.columns:
    print(df['retweet_count'].describe())
    # Remove extremely large retweet rows (example threshold: 10000)
    df = df[df['retweet_count'] <= 10000].reset_index(drop=True)
    print("After outlier filter rows:", df.shape[0])
```

```python
print("After cleaning - Rows, Columns:", df.shape)
print("\nMissing values per column (after):")
print(df.isnull().sum())
print("\nSample cleaned rows:")
display(df.head(10))
```

```python
clean_file = '/content/twitter_sentiment_200_cleaned.csv'
df.to_csv(clean_file, index=False)
print("Saved cleaned file:", clean_file)
```
