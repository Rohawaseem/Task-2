```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Step 1: Load dataset (✅ Replace with your actual file name)
df = pd.read_csv("twitter_sentiment_200.csv")

# Step 2: Check column names (optional)
print("Columns in dataset:", df.columns)

# Step 3: Clean text
df['clean_text'] = df['Tweet Text'].astype(str).str.lower()

# Step 4: Perform sentiment analysis
df['polarity'] = df['clean_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['calculated_sentiment'] = df['polarity'].apply(lambda p: 'positive' if p > 0 else ('negative' if p < 0 else 'neutral'))

# Step 5: Compare Sentiments (Bar chart)
sns.countplot(data=df, x='Sentiment')
plt.title('Original Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Step 6: Show calculated sentiment distribution
sns.countplot(data=df, x='calculated_sentiment')
plt.title('Calculated Sentiment Distribution (Using TextBlob)')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# 📊 Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Step 1️⃣: Load your dataset
df = pd.read_csv("twitter_sentiment_200.csv")

# Step 2️⃣: Basic cleanup
df['clean_text'] = df['Tweet Text'].astype(str).str.lower()

# Step 3️⃣: Sentiment polarity calculation
df['polarity'] = df['clean_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['calculated_sentiment'] = df['polarity'].apply(lambda p: 'positive' if p > 0 else ('negative' if p < 0 else 'neutral'))

# --- ✅ PLOT 1: Bar Chart – Original Sentiment Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sentiment', palette='Set2')
plt.title('Plot 1: Original Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# 💡 Insight:
# Most tweets belong to the "positive" category, showing users express more positivity overall.

# --- ✅ PLOT 2: Bar Chart – Calculated Sentiment Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='calculated_sentiment', palette='pastel')
plt.title('Plot 2: Calculated Sentiment Distribution (TextBlob)')
plt.xlabel('Calculated Sentiment')
plt.ylabel('Count')
plt.show()



# --- ✅ PLOT 3: Scatter Plot – Followers vs Favorites
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Followers', y='Favorites', hue='Sentiment', alpha=0.7)
plt.title('Plot 3: Relationship between Followers and Favorites')
plt.xlabel('Followers')
plt.ylabel('Favorites')
plt.show()

# 💡 Insight:
# Tweets from users with more followers tend to get more favorites, showing influence power.

# --- ✅ PLOT 4: Histogram – Tweet Length Distribution
df['tweet_length'] = df['Tweet Text'].apply(len)
plt.figure(figsize=(6,4))
plt.hist(df['tweet_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Plot 4: Tweet Length Distribution')
plt.xlabel('Tweet Length (characters)')
plt.ylabel('Frequency')
plt.show()


# --- ✅ PLOT 5: Box Plot – Retweets by Sentiment
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Sentiment', y='Retweets', palette='cool')
plt.title('Plot 5: Retweets Distribution across Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Retweets')
plt.show()

df.to_csv("twitter_sentiment_eda.csv", index=False)
print("✅ Week 3 completed! File 'twitter_sentiment_eda.csv' saved.")
```
