```python
import pandas as pd

# Upload your file from your computer
from google.colab import files
uploaded = files.upload()

# Load the dataset
df = pd.read_csv('twitter_sentiment_200.csv')

# Show first few rows
df.head()
```

```python
# Calculate correlation matrix
correlation = df[['Retweets', 'Favorites', 'Followers']].corrwith(df['Sentiment'].astype('category').cat.codes)
correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert Sentiment to numeric for visualization
df['SentimentCode'] = df['Sentiment'].astype('category').cat.codes

# Correlation heatmap
sns.heatmap(df[['Retweets', 'Favorites', 'Followers', 'SentimentCode']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Twitter Features")
plt.show()
```
